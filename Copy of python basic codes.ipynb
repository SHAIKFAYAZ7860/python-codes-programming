check given number is prime or not

```python
n=int(input("enter a prime number:"))
if n>1:
  for i in range(2,int(n/2)+1):
    if(n%i)==0:
        print("this is not a prime number ")
        break
    else:
      print("this is prime number")
  else:
    print("this is not prime number")
```

armstrong number

```python
num=int(input("enter a number:"))
sum=0
temp=num
while temp>0:
  digit=temp%10
  sum+=digit**3
  temp//=10
if num==sum:

  print("is armstrong number",num)
else:
  print("is not armstrong number",num)
```

fibonacci series

```python
n=int(input("enter a fibonacci number series u want?:"))
n1,n2=0,1
sum=0
count=1
print("fibonacci series",end="")
while count<n:
  print(sum,end="")
  n1=n2
  n2=sum
  sum=n1+n2
  count+=1
```

factorial number

```python


def factorial(n):
  if n==0:
    return 1
  return n*factorial(n-1)
num=int(input("enter a number:"))
print("The factorial of" ,num,"is",factorial(num))
```
